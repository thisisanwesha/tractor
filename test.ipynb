```python
test=pd.read_csv('OOT.csv')
test
```

```python
test.isnull().sum()
```

```python
sns.heatmap(test.isnull(), yticklabels=False)
```

```python
cust_id = test[['cust_id']]
cust_id
```

```python
test=test.drop(['cust_id'], axis=1)
```

```python
categorical_features=[feature for feature in test.columns if test[feature].dtypes=='O']
categorical_features
```

```python
test[categorical_features].isnull().sum()
```

```python
## no nan values for categorical variable
```

```python
test = pd.get_dummies(test, columns = ['NEVER_BOUNCE_FLAG'],drop_first=True) 
test
```

```python
test=test.drop(['TOTAL_OUTSTANDING_BALANCE LIVE_TRACTOR LOAN'], axis=1)
```

```python
numerical_nan=[feature for feature in test.columns if test[feature].isnull().sum()>1 and test[feature].dtypes!='O']
```

```python
for feature in numerical_nan:
    ## create a new feature to capture nan values
    test[feature+'nan']=np.where(test[feature].isnull(),1,0)
    ## We will replace by using median
    median_value=test[feature].median()
    test[feature].fillna(median_value,inplace=True)
    
test[numerical_nan].isnull().sum()
```

```python
test
```

```python
test.isnull().sum()
```

```python
feature_scale=[feature for feature in test.columns]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(test[feature_scale])
```

```python
scaler.transform(test[feature_scale])
```

```python
final= pd.DataFrame(scaler.transform(test[feature_scale]), columns=feature_scale)
```

```python
cust_id = pd.DataFrame(cust_id)
cust_id
```

```python
processed_test = pd.concat([cust_id,final], axis= 1)
processed_test
```

```python
processed_test.to_csv('processed_test.csv',index=False)
```
